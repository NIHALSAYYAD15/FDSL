8. Null values: - Handling null values using pandasâ€™ DataFrame

```python
#Handling NULL values
#isnull()
import pandas as pd
import numpy as np
```

```python
# dictionary of lists
dict = {'First Score':[100, 90, np.nan, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score':[np.nan, 40, 80, 98]}
# creating a dataframe from list
df = pd.DataFrame(dict)
```

```python
# creating a dataframe from list
df = pd.DataFrame(dict)
```

```python
# using isnull() function  
df.isnull()
```

```python
#df.notnull()
df.fillna(0)
```

```python
#Fillnull values with the previous ones
df.fillna(method ='pad')
```

```python
#Fillnull values with the afters ones
df.fillna(method ='bfill')
```

```python
#replace()
df.replace(to_replace = np.nan, value = -99)
```

```python
#Using interpolate() function to fill the missing values using linear method.
df.interpolate(method ='linear', limit_direction ='forward')
```

```python
#dropna()
df.dropna()
```

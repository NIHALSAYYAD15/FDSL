5. Indexing & Slicing: - Indexing & slicing operations on 2 D NumPy array.

```python
import numpy as np
a=np.array([[1,2,3],[4,5,6]])     #2-D Arrays
```

```python
print(a)
```

```python
print("Each item contains",a.itemsize,"bytes")
```

```python
#The type of array.
print(type(a))
```

```python
#Numbers of Dimensions.
print(a.ndim)
```

```python
#Array Indexing.
print("The value is=",a[1][2])
print("The value is=",a[0])
print("The value is=",a[-2])    #Negative Indexing
print("The value is=",a[-1][-2])
```

```python
#Array Slicing
a=a+1;
print(a)
print(a.sum())
print(a[1,1:2])
print(a[0:1,1])
print(a[:,:-1])
print(a[:,:1])
```

```python
arr1 = np.array([[1, 2, 3, 4, 5],             #2 D Array
                 [6, 7, 8, 9, 10]])
print(arr1[0,1:4])
print(arr1[1,1:4])

print(arr1[0:2,1:4])

print(arr1[0:2,3:6])
```

13 EDA (Removing Outliers)- Detecting & removing outlier using DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
# Generate sample data
X, y = make_blobs(n_samples=30, centers=3, n_features=2, random_state=42)
# Create DBSCAN object
dbscan = DBSCAN(eps=1.5, min_samples=4)
# Fit model and predict cluster labels
labels = dbscan.fit_predict(X)
# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow')
plt.show()
print("Number of outliers:", sum(labels==-1))
```

```python
# Identify outliers
outliers = X[labels == -1]

# Remove outliers
X_clean = X[labels != -1]

# Plot the clusters without outliers
plt.scatter(X_clean[:, 0], X_clean[:, 1], c=labels[labels != -1], cmap='rainbow')
#plt.scatter(outliers[:, 0], outliers[:, 1], color='red', marker='x', label='Outliers')
plt.legend()
plt.show()
```

10. Matplotlib (3 D): - 3D & sub plotting in matplotlib
(Counter, Surface, Scatter, Line, Density)

```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
#create data for plot
z=np.linspace(0,np.pi*2)
x=np.sin(z)
y=np.cos(z)
ax=plt.figure().add_subplot(projection='3d')
ax.plot(x,y,z)
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z-axis')
plt.show()
```

```python
# create data for the scatter plot
x = np.random.normal(size=100)
y = np.random.normal(size=100)
z = np.random.normal(size=100)
colors = np.random.rand(100)
# create a 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# plot the scatter plot
ax.scatter(x, y, z, c=colors, cmap='viridis')
# set the labels for the axes
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
```

```python
# create data for the density and contour plot
x = np.random.normal(size=1000)
y = np.random.normal(size=1000)
z = (x**2+y**2)/np.pi
#z = (x**2+y**2)
# create a 3D figure
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
# plot the density and contour plot
hist, xedges, yedges = np.histogram2d(x, y, bins=30, density=True)
X, Y = np.meshgrid(xedges[:-1], yedges[:-1])
Z = hist.T
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax.contour(X, Y, Z, offset=-0.15, cmap='viridis')
# set the labels for the axes
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
#show plot
plt.show()
```

```python
# importing mplot3d toolkits, numpy and matplotlib
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()

# syntax for 3-D projection
ax = plt.axes(projection ='3d')

# defining all 3 axis
z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)

# plotting
ax.plot3D(x, y, z, 'green')
ax.set_title('3D line plot geeks for geeks')
plt.show()
```
